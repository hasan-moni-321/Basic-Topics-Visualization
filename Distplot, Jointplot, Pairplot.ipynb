```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.dtypes
```

```python
#finding correlation
correlation = tips.corr()
correlation
```

```python
#correlation in heatmap
sns.heatmap(correlation, annot=True)
```

# jointplot

```python
#jointplot with kind='hex'
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

most of the total bill between 10 to 20 doller. That is also clear in hist plot. When total bill increase 
amount of tips also increase. There are some outliers in tips  

```python
#joint plot with regression
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

# pairplot

```python
sns.pairplot(tips)
```

```python
#pairplot with hue='sex'
sns.pairplot(tips, hue='sex')
```

```python
#counting smokers and non-smokers
tips['smoker'].value_counts()
```

most of the customers are not smoker
