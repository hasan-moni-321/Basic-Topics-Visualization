```python
sns.relplot('total_bill', 'tip', data=tips)
```

```python
sns.relplot('total_bill','tip', hue='time', data=tips)
```

```python
sns.relplot('total_bill', 'tip', hue='day', col='time', data=tips)
```

```python
sns.relplot('total_bill','tip', hue='day', col='time', row='sex', data=tips)
```

```python
sns.relplot('total_bill', 'tip', hue='time', col='day', data=tips )
```

```python
sns.relplot('total_bill', 'tip', hue='time', col_wrap=2, col='day', data=tips)
```

```python
sns.relplot('total_bill', 'tip', hue='time', size='size',col='time', data=tips)
```
