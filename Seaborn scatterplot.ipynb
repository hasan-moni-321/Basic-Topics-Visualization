```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
a = tips.sex.unique()
print(a)
b = tips.smoker.unique()
print(b)
c = tips.day.unique()
print(c)
```

```python
sns.scatterplot('total_bill', 'tip', data=tips)
```

```python
sns.scatterplot('total_bill', 'tip', hue='time', data=tips)
```

```python
sns.scatterplot('total_bill', 'tip', hue='time', style='time', data=tips)
```

```python
sns.scatterplot('total_bill','tip', hue='day', style='time', data=tips)
```

```python
sns.scatterplot('total_bill','tip', size='size', data=tips)
```

```python
sns.scatterplot('total_bill', 'tip', hue='size', size='size', data=tips)
```

```python
marker = {'Lunch':'s', 'Dinner':'X'}

sns.scatterplot(x='total_bill', y='tip', style='time', markers=marker, data=tips)
```

```python
sns.scatterplot('total_bill','tip', marker='+', s=100, color='.2', data=tips)
```
